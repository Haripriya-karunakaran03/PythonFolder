```
In python, the file system contains the files and directories. To handle these files and directories python supports os module. Python has the os module, which provides us with many useful methods to work with directories (and files as well).

The os module provides us the methods that are involved in file processing operations and directory processing like renaming, deleting, get current directory, changing directory etc.
```

````python
# Renaming the file - rename()
# Syntax - os.rename (current-name, new-name)

import os
os.rename("x2.txt","x3.txt")
````

````python
# Creating the new directory  mkdir()
# The mkdir() method is used to create the directories in the current working directory.

# Syntax - os.mkdir(directory-name)

import os
os.mkdir("Haripriya")
````

````python
# Changing the current working directory  chdir()
# The chdir() method is used to change the current working directory to a specified directory.

# Syntax - os.chdir("new-directory")

os.chdir("Desktop")
````

````python
# os.walk()! It's a handy function in Python's os module for traversing directories. 
# You can use it to iterate through all the files and subdirectories within a given directory
import os
current_dir = "Pictures"
for dir_path,dir_names,file_name in os.walk(current_dir):
    for j in file_name:
        print(j)
````

### difference between r+ and w+
```
w+ opens the file to write and read both. 
It is different from r+ in the sense that it overwrites the previous file if one exists whereas r+ doesn't overwrite 
the previously written file. It creates a new file if no file exists. The file pointer exists at the beginning of the file
```
